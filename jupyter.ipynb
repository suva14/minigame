```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_files = [
    r"C:\Users\lysuv\OneDrive - De Vinci\A4 Createch\minigame\daniel.csv",
    r"C:\Users\lysuv\OneDrive - De Vinci\A4 Createch\minigame\suva.csv",
    r"C:\Users\lysuv\OneDrive - De Vinci\A4 Createch\minigame\coco.csv",
    r"C:\Users\lysuv\OneDrive - De Vinci\A4 Createch\minigame\loezig.csv",
    r"C:\Users\lysuv\OneDrive - De Vinci\A4 Createch\minigame\come.csv"
]
```

```python
dataframes = []
for file in csv_files:
    try:
        df = pd.read_csv(file)
        df['Player'] = file.split('.')[0]  # Ajouter le nom du joueur depuis le nom du fichier
        dataframes.append(df)
    except FileNotFoundError:
        print(f"Fichier {file} non trouv√©.")
```

```python
combined_data = pd.concat(dataframes, ignore_index=True)
combined_data['Date'] = pd.to_datetime(combined_data['Date'])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(combined_data['Error Count'], combined_data['Score'], alpha=0.7, c='blue', edgecolors='k')
plt.title("Score vs. Nombre d'Erreurs", fontsize=14)
plt.xlabel("Nombre d'Erreurs", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(combined_data['Game Time (seconds)'], combined_data['Score'], alpha=0.7, c='green', edgecolors='k')
plt.title("Temps de jeu vs. Score", fontsize=14)
plt.xlabel("Temps de jeu (secondes)", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(combined_data['Score'], bins=10, color='orange', edgecolor='black', alpha=0.7)
plt.title("Distribution des Scores", fontsize=14)
plt.xlabel("Score", fontsize=12)
plt.ylabel("Fr√©quence", fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```
